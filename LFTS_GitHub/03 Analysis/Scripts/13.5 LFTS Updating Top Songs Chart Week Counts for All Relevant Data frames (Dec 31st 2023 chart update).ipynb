```python
# Update total top songs for albums

# Create a for loop to update top songs

alb_top_s_count= []

for value in df['album_id']:
    if value == "1989":
        alb_top_s_count.append(10)
    elif value == "DEBUT":
        alb_top_s_count.append(5)
    elif value == "REP":
        alb_top_s_count.append(6)
    elif value == "FEAR":
        alb_top_s_count.append(16)
    elif value == "LOVE":
        alb_top_s_count.append(18)
    elif value == "SPEAK":
        alb_top_s_count.append(17)
    elif value == "RED":
        alb_top_s_count.append(11)
    elif value == "FOLK":
        alb_top_s_count.append(16)
    elif value == "EVER":
        alb_top_s_count.append(15)
    elif value == "FEARTV":
        alb_top_s_count.append(8)
    elif value == "REDTV":
        alb_top_s_count.append(26)
    elif value == "MIDN":
        alb_top_s_count.append(22)
    elif value == "SPEAKTV":
        alb_top_s_count.append(22)
    elif value == "1989TV":
        alb_top_s_count.append(21)
    else:
        alb_top_s_count.append("unknown")
```

```python
# update with new values

df['top_s_cnt'] = alb_top_s_count
```
