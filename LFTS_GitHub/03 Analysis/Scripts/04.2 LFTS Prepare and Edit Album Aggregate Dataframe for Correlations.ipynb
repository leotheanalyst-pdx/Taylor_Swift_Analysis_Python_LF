# Tracks

```python
tracks = []

for value in alb_sales['album_id']:
    if value == "1989":
        tracks.append(19)
    elif value == "1989TV":
        tracks.append(22)
    elif value == "DEBUT":
        tracks.append(15)
    elif value == "EVER":
        tracks.append(17)
    elif value == "FEAR":
        tracks.append(19)
    elif value == "FEARTV":
        tracks.append(26)
    elif value == "FOLK":
        tracks.append(17)
    elif value == "LOVE":
        tracks.append(18)
    elif value == "MIDN":
        tracks.append(23)
    elif value == "RED":
        tracks.append(22)
    elif value == "REDTV":
        tracks.append(30)
    elif value == "REP":
        tracks.append(15)
    elif value == "SPEAK":
        tracks.append(20)
    elif value == "SPEAKTV":
        tracks.append(22)
    else:
        tracks.append('unknown')
```

```python
alb_sales['track_cnt'] = tracks
```
