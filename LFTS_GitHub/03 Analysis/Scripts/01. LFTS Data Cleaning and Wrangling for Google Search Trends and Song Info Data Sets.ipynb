Observations: Successfully renamed Taylor Swift album to be Debut

```python
# Create new column for album id

result = []
for value in df_song_info_stud_alb['album']:
    if value == "Red (Taylor's Version)":
        result.append("REDTV")
    elif value == "Midnights (The Til Dawn Edition)":
        result.append("MIDN")
    elif value == "Red (Deluxe Edition)":
        result.append("RED")
    elif value == "Fearless (Taylor's Version)":
        result.append("FEARTV")
    elif value == "1989 (Taylor's Version) [Deluxe]":
        result.append("1989TV")        
    elif value == "Speak Now (Taylor's Version)":
        result.append("SPEAKTV")
    elif value == "Red (Deluxe Edition)":
        result.append("RED")
    elif value == "Speak Now (Deluxe Edition)":
        result.append("SPEAK")
    elif value == "1989 (Deluxe Edition)":
        result.append("1989")
    elif value == "Fearless Platinum Edition":
        result.append("FEAR")
    elif value == "Lover":
        result.append("LOVE")
    elif value == "evermore (deluxe version)":
        result.append("EVER")
    elif value == "folklore (deluxe version)":
        result.append("FOLK")
    elif value == "reputation":
        result.append("REP")
    elif value == "Debut":
        result.append("DEBUT")
    else:
        result.append("Unknown")
```

```python
df_song_info_stud_alb['album_id'] = result
```

```python
df_song_info_stud_alb['album_id'].value_counts(ascending=False)
```

```python
# Crosstab of albums, by name, and song count to check with album_id counts

crosstab_albums = df_song_info_stud_alb.groupby('album').agg({'song_name':['count']}, dropna=False, ascending =False)
```

```python
crosstab_albums
```

Song count values match album counts for both "album" and actually tracks.
