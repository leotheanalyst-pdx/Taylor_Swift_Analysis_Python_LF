```python
# Create a subset to focus on for seaborn correlation heatmap

album_sub = df_alb_num[['alb_num', 'wk1_alb_sales', 'tot_alb_plays', 'avg_daily_alb_streams', 'top_s_cnt', 
                        'top_alb_count', 'alb_r_day', 'dance_m', 'speech_m']]
```

```python
# Create heatmap, highlighting key relationships (or lackthereof)

f,ax = plt.subplots(figsize=(9, 9))
cmap = sns.diverging_palette(220, 20, as_cmap=True) # Set diverging color palette
album_corrmat= sns.heatmap(album_sub.corr(), cmap=cmap, annot=True, vmin=-1, vmax=1, ax=ax) 

# Add title to the heatmap
album_corrmat.set_title("Correlation Heatmap of Key Album Measures")

# Adjust layout to make sure the entire figure is visible
f.tight_layout()
```

```python
album_corrmat.figure.savefig(os.path.join(path, '03 Analysis', 'Visualizations',  'Dashboard Python Visuals', 'album_corrmat.png'))
```

# Billboard Top 200 Albums Dataframe
### Includes: 
- Chart week, position, peak position, weeks on chart, album information including all aggregates

- Need to quickly add in album cluster

File name: bill_alb_aggs_merge.csv

```python
df_bill_albs.head()
```

```python
df_bill_albs.tail()
```

```python
# Create for loop to insert album cluster into full billboard dataframe
# Will assign album clusters (from k-means cluster analysis) to each album using album_id

album_cluster_number = []

for value in df_bill_albs['album_id']:
    if value == "1989":
        album_cluster_number.append("1")
    elif value == "DEBUT":
        album_cluster_number.append("3")
    elif value == "REP":
        album_cluster_number.append("1")
    elif value == "FEAR":
        album_cluster_number.append("3")
    elif value == "LOVE":
        album_cluster_number.append("1")
    elif value == "SPEAK":
        album_cluster_number.append("3")
    elif value == "RED":
        album_cluster_number.append("1")
    elif value == "FOLK":
        album_cluster_number.append("2")
    elif value == "EVER":
        album_cluster_number.append("2")
    elif value == "REDTV":
        album_cluster_number.append("4")
    elif value == "MIDN":
        album_cluster_number.append("2")
    elif value == "SPEAKTV":
        album_cluster_number.append("4")
    elif value == "1989TV":
        album_cluster_number.append("4")
    else:
        album_cluster_number.append("unknown")
```

```python
df_bill_albs['alb_cluster_num'] = album_cluster_number
```

```python
df_bill_albs['alb_cluster_num'].value_counts(dropna=False)
```
